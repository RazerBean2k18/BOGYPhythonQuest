```python
import math
error_var = False
def is_number(string):
    try:
        float(string)
        return True
    except ValueError:
        return False
    
def error_message(errror_var):
    print("Es ist ein Fehler aufgetreten!")
```

```python
operation = input("Welcher Rechenart soll genutzt werden?\n\n> Addition \t\t(+)\n> Subtraktion \t\t(-)\n> Multiplikation \t(*)\n> Division \t\t(/)\n> Wurzel ziehen \t(I)\n> Sinus berechnen \t(II)\n> Cosinus berechnen \t(III)\n> Tangens berechnen \t(IV)\n> Modulo berechnen \t(V)\n\n")
operation = operation.lower()

if operation == "addition" or operation == "+":
    print("\nSie haben Addition gewählt.")
    operator = 1

elif operation == "subtraktion" or operation == "-":
    print("\nSie haben Subtraktion gewählt.")
    operator = 2

elif operation == "multiplikation" or operation == "*":
    print("\nSie haben Multiplikation gewählt.")
    operator = 3

elif operation == "division" or operation == "/":
    print("\nSie haben Division gewählt.")
    operator = 4
    
elif operation == "wurzel ziehen" or operation == "i":
    print("\nSie haben Wurzel ziehen gewählt.")
    operator = 5
    
elif operation == "sinus berechnen" or operation == "ii":
    print("\nSie haben Sinus berechnen gewählt.")
    operator = 6
    
elif operation == "cosinus berechnen" or operation == "iii":
    print("\nSie haben Cosinus berechnen gewählt.")
    operator = 7
    
elif operation == "tangens berechnen" or operation == "iv":
    print("\nSie haben Tangens berechnen gewählt.")
    operator = 8
    
elif operation == "modulo berechnen" or operation == "v":
    print("\nSie haben Modulo berechnen gewählt.")
    operator = 9

else:
    print("\nEs ist etwas schief glaufen!")
```

```python
numbers = []
x = 0
if operator == 1 or operator == 2 or operator == 3:
    while x < 2:
        new = input("\nGeben Sie eine Nummer an:\n")
        new = float(new)
        if is_number(new) == False:
            if new != "-":
                error_message()
                error_var = True
                break
            else:
                break 
        if is_number(new) == True:
            numbers.append(new)
elif operator == 4:
    while x < 2:
        new = input("\nGeben Sie eine Nummer an:\n")
        new = float(new)
        if is_number(new) == False:
            if new == "-":
                break
            else:
                error_message()
                error_var = True
                break 
        elif is_number(new) == True:
            if new == 0:
                error_message()
                error_var = True
                print("Die Division ist nicht durch 0 möglich!")
                break
            else:
                numbers.append(new)
                
elif operator == 9:
    while x < 2:
        new = input("\nGeben Sie eine Nummer an:\n")
        new = float(new)
        if is_number(new) == False:
            if new == "-":
                break
            else:
                error_message()
                error_var = True
                break 
        elif is_number(new) == True:
            if new == 0:
                error_message()
                error_var = True
                print("Die Division ist nicht durch 0 möglich!")
                break
            else:
                numbers.append(new)
        x = x + 1
                
else:
    new = input("\nGeben Sie eine Nummer an:\n")
    new = float(new)
    numbers.append(new)
```

```python
if error_var == False:
    result = float(numbers[0])

    if operator == 1:
        for number in numbers:
            number = float(number)
            result = result + number
        result = result - float(numbers[0])

    elif operator == 2:
        for number in numbers:
            number = float(number)
            result = result - number
        result = result + float(numbers[0])
        
    elif operator == 3:
        for number in numbers:
            number = float(number)
            result = result * number
        result = result / float(numbers[0])
    
    elif operator == 4:
        for number in numbers:
            number = float(number)
            result = result / number
            
    elif operator == 5:
        result = math.sqrt(numbers[0])
    
    elif operator == 6:
        result = math.sin(math.radians(numbers[0]))
        
    elif operator == 7:
        result = math.cos(math.radians(numbers[0]))
        
    elif operator == 8:
        result = math.tan(math.radians(numbers[0]))
     
    elif operator != 9:
        result = str(result)
        print("Das Ergebnis ist: " + result + "!")
    
    
    elif operator == 9:
        pre_result = numbers[0] / numbers[1]
        pre_result = int(pre_result)
        result = numbers[0] % numbers[1]
        pre_result = str(pre_result)
        result = str(int(result))
        print("Das Ergebnis ist " + pre_result + " Rest: " + result + "!")
    
else:
    print("\nEs ist etwas schief glaufen, die Rechnung konnte nicht vollzogen werden!")
```
